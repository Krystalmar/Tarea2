La solución de la ecuación de Laplace
$$\frac{\partial u^2}{\partial x^2} = \frac{\partial^2 u}{\partial y^2} \;\;\; 0<x<1; \;\; 0<y<1$$
sujeta a las condiciones de frontera:
$$u(0,y)=u(1,y)=0$$
Y la condición inicial:
$$u(x,0)=100,\;\;\; u(x,1)=200$$
Esta dada en términos de la serie de Fourier por la expresión:
$$u(x,y)=\sum_{n=1}^{\infty}sen(nπx)\left[200\left(\frac{1-(-1^n)}{nπ}\right)cosh(nπy)+senh(nπy)\left[\frac{1-(-1^n)}{nπ}⋅200\left(\frac{2-cosh(nπ)}{senh(nπ)}\right)\right]\right]$$

```python
from pylab import *
```

```python
def u_xy(x,y,N=30):
    f=zeros((len(x),len(y)))
    for i in range(len(y)):
      for n in range (1,N):
        a=200*(1-(-1)**n)/(n*pi)
        b=a*(2-cosh(n*pi))/sinh(n*pi)
        f[:,i] += sin(n*pi*x)*((a*cosh(n*pi*y[i]))+(b*sinh(n*pi*y[i])))
    return f
```

```python
x = linspace(0,1,1000)
y = linspace(0,1,1000)
U = u_xy(x,y)
```

```python
plot(x,U[:,0],"darkred",label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-0.01,120.5]),grid(),legend()
show()
```

Vemos cómo la solución evoluciona cuando $y \to \infty$.

```python
plot(x,U[:,0],label='$u(x,y=0)$') 
plot(x,U[:,1],label='$u(x,y=1)$')
plot(x,U[:,2],label='$u(x,y=2)$')
plot(x,U[:,3],label='$u(x,y=3)$')
plot(x,U[:,10],label='$u(x,y=10)$')
plot(x,U[:,20],label='$u(x,y=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,y)$'); 
plt.title("Evolución de la solución")
xlim([-0.01,1.01]),ylim([-0.30,120.5]),grid(),legend()
show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
y = linspace(0,1.0,100)

X,Y = np.meshgrid(x,y)

U = u_xy(x,y)

plt.xlabel('$y$'); plt.ylabel('$x$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="magma")
```
